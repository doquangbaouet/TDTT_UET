## W6A19 
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list


```python
def index_max(ds):
   max_temp = 0 
   for i in range(len(ds)):
      if ds[i] > ds[max_temp]:
         max_temp = i 
   return max_temp 
a = [1,3,8,7,5]
print("vi tri lon nhat la:", index_max(a))
```

## W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1



```python
def tim_vi_tri_k(ds, k):
    index = 0 
    for so in ds:
        if so == k:
            return index 
        index += 1 
    return -1 
a = list(map(int, input().split()))
b = input()
res = tim_vi_tri_k(a, b)
print("vi tri la:", res)
```

## W6A21
 Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện


```python
# Bước 1: Nhập vào các chuỗi và tách ra thành danh sách
danh_sach_chuoi = input().split()

# Bước 2: Chuẩn bị "cuốn sổ" (Dictionary) để ghi chép
tu_dien_vi_tri = {}

# Bước 3: Dùng "cái chân" index để đi tuần tra
index = 0
for chuoi in danh_sach_chuoi:
    # Nếu cái tên này chưa có trong sổ, mình tạo một cái list rỗng cho nó
    if chuoi not in tu_dien_vi_tri:
        tu_dien_vi_tri[chuoi] = []
    
    # Ghi lại vị trí hiện tại của chuỗi đó vào list
    tu_dien_vi_tri[chuoi].append(index)
    
    # Bước tiếp sang vị trí sau
    index = index + 1

# Bước 4: Chuyển các list vị trí thành dạng tuple như đề bài yêu cầu
for chuoi in tu_dien_vi_tri:
    tu_dien_vi_tri[chuoi] = tuple(tu_dien_vi_tri[chuoi])

# In kết quả cuối cùng
print(tu_dien_vi_tri)
```
