## W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 


```python
import math
n = int(input("nhap so n:"))
k = math.log(n,2)
if k == int(k):
    print(f"{n} co dang 2^k")
else:
    print(f"{n} khong co dang 2^k")
```

## W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("nhap so n:"))
tongchuso = 0
while n > 0:
    so_cuoi = n % 10 
    tongchuso += so_cuoi
    n //= 10
print(tongchuso)

```

## W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input("nhap so n:"))
so_chan = 0
so_le = 0
while n > 0: 
    so_cuoi = n % 10 
    if so_cuoi % 2 == 0:
        so_chan += 1
    else:
        so_le += 1
    n //= 10 
print(f"so so chan {so_chan}")
print(f"so so le {so_le}")
```

## W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
n = int(input("nhap so n:"))
if n == -1:
    print("nguoi dung ngung nhap")
else:
   max_val = n
   min_val = n
   n = int(input("nhap so n:"))
   while n != -1:
       if n > max_val:
           max_val = n
       if n < min_val:
           min_val = n
       n = int(input("nhap so n:"))
   print(f"so lon nhat la {max_val}")
   print(f"so nho nhat la {min_val}")
```
