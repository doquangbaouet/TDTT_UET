không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25….

## Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
n = int(input("nhap so n:"))
if n == -1:
    print("dung nhap so")

else:
    max_value = n
    min_value = n 
    n = int(input("nhap so n: "))
    while n != -1:
        if n > max_value:
            max_value = n 
        if n < min_value:
            min_value = n 
        n = int(input("nhap so n:"))
    print(f"so lon nhat la {max_value}")
    print(f"so nho nhat la {min_value}")
   



```

## Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ 

```python
n = int(input("nhap so m:"))
so_chan = 0 
so_le = 0
while n > 0: 
    so_cuoi = n % 10 
    if so_cuoi % 2 == 0: 
        so_chan += 1 
    else:
        so_le += 1 
    n = n // 10
print(f"so chan {so_chan}")
print(f"so so le {so_le}")


```

## Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
a = int(input("nhap so a:"))
a1 = 1
a2 = 1
if a <= 1: 
    print("khong co  so fibonacci phu hop")
else:
    while a1 + a2 <= a:
        a3 = a1 + a2
        a1 = a2
        a2 = a3 
    print(" so fibonacci phu hop la:", a2)
```

##Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 

```python
n = int(input("nhap so n:"))
so_so_nt = 0
while n > 0:
    so_cuoi = n % 10 
    if so_cuoi == 2 or so_cuoi == 3 or so_cuoi == 5 or so_cuoi == 7:
        so_so_nt += 1
    n //= 10
print("so chu so nguyen to la:", so_so_nt)
```
