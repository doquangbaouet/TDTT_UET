## W5A16
Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def kiem_tra_amstrong(x):
    n = len(str(x)) ## số mũ tương đương với số số 
    temp = x 
    tong = 0 
    while temp > 0:
        chu_so = temp % 10 
        tong += chu_so**n 
        temp //= 10 
    return tong == x 

x = int(input())
if kiem_tra_amstrong(x):
    print("True")
else:
    print("False")
  
    
    
```

## W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.



```python
def kiem_tra_so_ky_dieu(n):
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong == n
n = int(input())
if kiem_tra_so_ky_dieu(n):
    print("True")
else:
    print("False")
```

## W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def quadratic_roots_count(a, b, c):
    delta = b**2 - 4 * a * c
    if delta > 0:
        x1 = (-b + delta**0.5) / (2 * a)
        x2 = (-b - delta**0.5) / (2 * a)
        return x1, x2
    elif delta == 0:
        x = -b / (2 * a)
        return x
    else:
        return 'phuong trinh khong co nghiem thuc'
print(quadratic_roots_count(1,4,4))
print(quadratic_roots_count(1,1,1))
print(quadratic_roots_count(2,3,1))


```

## W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).
## BA =(X1-X2,Y1-Y2) VÀ BC = (X3-X2,Y3-Y2)

```python
import math 
def tinh_goc_ABC(x1,y1,x2,y2,x3,y3):
    cos_ABC = ((x2-x1)*(x3-x1) + (y2-y1)*(y3-y1)) / (math.sqrt((x2-x1)**2 + (y2-y1)**2) * math.sqrt((x3-x1)**2 + (y3-y1)**2))
    goc_ABC = math.acos(cos_ABC)
    return math.degrees(goc_ABC) 
x1,y1,x2,y2,x3,y3 = map(float, input().split())
print(tinh_goc_ABC(x1,y1,x2,y2,x3,y3))
```

## W5A30 Viết hàm tính hệ số nhị thức C(n, k).
C(k,n) = n! / k!.(n-k!)

```python
import math 
def to_chop_chap_k_cua_n(k,n):
    return math.factorial(n) / (math.factorial(k)*math.factorial(n-k))
print(int(to_chop_chap_k_cua_n(4,10)))
```
