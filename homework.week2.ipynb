Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a, b, c = map(float, input().split())
if a + b > c and a + c > b and b + c > a : 
    if a == b == c : 
        print("tam giac deu")
    elif a == b or b == c or c == a :
        print("tam giac can")
    elif a**2 == b**2 + c**2 or b**2 == a **2 + c**2 or c**2 == b**2 + a**2 :
        print("tam giac vuong")
    else: print("tam giac thuong")
    p = (a + b + c)/2 
    chu_vi = a + b + c 
    dien_tich = (p*(p-a)*(p-b)*(p-c))**0.5
    print(f" chu vi la: {chu_vi}")
    print(f" dien tich la: {dien_tich}")
else: 
    print("Khong la ba canh cua mot tam giac")



```

## Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(float, input().split())
delta = b**2 -4 *a *c 
if delta > 0: 
    x1 = (-b +(b**2-4*a*c)**0.5)/(a*2) 
    x2 = (-b -(b**2-4*a*c)**0.5)/(a*2)
    print(f"nghiem cua phuong trinh bac 2 la: {x1} va {x2}")
elif delta == 0 : 
    x = -b/(a*2)
    print(f"phuong trinh bac 2 co nghiem kep la: {x}")
elif delta < 0 :
    print(" phuong trinh bac 2 vo nghiem ")
```

## Bài 8: Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n

```python
a, b, m, c, d, n = map(float, input().split())
# su dung he Crammer 
D = a*d - b*c
Dx = m*d -b*n 
Dy = a*n -c*m
if D != 0 :
    x = Dx / D 
    y = Dy / D 
    print(f"he phuong trinh co nghiem la: {x} va {y}")
elif D == 0:
    if Dx == 0 and Dy == 0: 
        print("he phuong trinh co vo so nghiem")
    if Dx != 0 or Dy != 0 :
        print("he phuong trinh vo nghiem")
```

## Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây. 

```python
x = float(input())
gio = x // 3600
giay_luc_sau = x % 3600
phut = giay_luc_sau // 60
giay_final = giay_luc_sau % 60 
print(f"{gio} gio, {phut} phut, {giay_final} giay")
```

## Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
# Tam I(a,b) ban kinh R , toa do diem A(x,y)
a, b, R, x, y = map(float, input().split())
IA = ((x-a)**2+(y-b)**2)**0.5
if IA == R :
    print("diem A nam tren duong tron")
if IA < R :
    print("diem A thuoc duong tron") 
if IA > R :
    print("diem A nam ngoai duong tron")

```
